```python
def search(a, v):
    """
    `a` must be in ascending order
    
    find an `i` such that `a[i-1] < v <= a[i]`
    
    maybe return number of comparisons?
    """

    if len(a) == 0:
        return 0
    
    i = 0
    j = len(a)
    
    comparisons = 0

    if v <= a[0]:
        return 0, comparisons
    if a[j-1] < v:
        return j, comparisons
    
    i += 1
    
    # otherwise: a[0] < v < a[j]
    
    while i != j:
        k = i + (j-i)//2
        
        print(i,k,j)
        
        if v <= a[k]:
            comparisons += 1
            if j != k:
                j = k                
                continue

        if a[k-1] < v:
            comparisons += 1
            if i != k:
                i = k
                continue

    return i, comparisons
```

```python
import numpy as np
import matplotlib.pyplot as plt


def make_prob():
    N = 400

    a = sorted(np.random.randint(N//3, size=N))
    v = np.random.randint(N) - N//3

    #a[0], a[-1], len(a), v
    
    return a,v

def check_sol(a,v,i):
    if i == 0:
        assert v <= a[0]
    elif i == len(a):
        assert a[-1] < v
    else:
        assert a[i-1] < v <= a[i]
```

```python
def foo(N=100):
    for _ in range(N):
        a, v = make_prob()
        i, comp = search(a, v)
        check_sol(a,v,i)
        yield comp
```

```python
np.mean(list(foo(10000)))
```
