```python
import numpy as np
import matplotlib.pyplot as plt
import cvxpy as cvx

eps = 1

x = np.linspace(0,eps,1000)
y = x**2
```

```python
fig, ax = plt.subplots(figsize=(5,5))

ax.plot(x,y, alpha=.5)
ax.axhline(0, color='k', alpha=.3)
ax.axvline(0, color='k', alpha=.3)
ax.grid(alpha=.2)
ax.set_aspect(1)
```

```python
a = np.linspace(0,2*eps,1000)
b = a**2

b = b - 2*a*eps + eps**2
a = np.abs(a - eps)
```

```python
fig, ax = plt.subplots(figsize=(5,5))

ax.plot(x,y)
ax.plot(a,b, alpha=.5)
ax.axhline(0, color='k', alpha=.3)
ax.axvline(0, color='k', alpha=.3)
ax.grid(alpha=.2)
ax.set_aspect(1)
```
