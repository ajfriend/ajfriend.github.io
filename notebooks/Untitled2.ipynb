```python
x = np.linspace(-7, 2, 1000)

theta = np.pi/4 - .14

y = np.exp(x) - .8

R = np.array([
    [np.cos(theta), np.sin(theta)],
    [-np.sin(theta), np.cos(theta)],
])




z = R@np.array([x,y])
```

```python
fig, ax = plt.subplots(figsize=(5,5))

ax.plot(z[0],z[1], alpha=.5)
ax.plot(-z[0],z[1],alpha=.5 )
ax.axhline(0, color='k', alpha=.3)
ax.axvline(0, color='k', alpha=.3)
ax.grid(alpha=.2)
```
