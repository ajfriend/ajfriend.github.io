## also

- can do approximations around q == 1 and q == 0 (add both!)
- that'll give us the best endpoint behavior

````python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(.001, 1, 1000)

ys = [
    1/x - 1,
    np.exp(x) + np.exp(-x) - 2,
    x**2,
    np.exp(x) - 1,
]

ys = [
    np.clip(y, 0, 1)
    for y in ys
]


for y in ys:
    plt.plot(x,y)
````
